```python
import pandas as pd

df = pd.read_csv("Life Expectancy Data.csv")
```

```python
print(df.columns)
```

#1.How many unique countries are present in the dataset?

```python
unique_countries = df['Country'].nunique()
print("Number of unique countries:", unique_countries)
```

#2.What is the range of years covered in the dataset?

```python
min_year = df['Year'].min()
max_year = df['Year'].max()
print("Range of years:", min_year, "-", max_year)
```

#3.How many records are there for each country?

```python
records_per_country = df['Country'].value_counts()
print("Records per country:")
print(records_per_country)
```

#4.What is the average life expectancy across all countries and years?

```python

average_life_expectancy = df['Life expectancy'].mean()
print("Average life expectancy:", average_life_expectancy)
```

#5.Which country has the highest life expectancy, and in which year was it recorded?

```python
max_life_expectancy = df.loc[df['Life expectancy'].idxmax()]
country = max_life_expectancy['Country']
year = max_life_expectancy['Year']
print("Country with the highest life expectancy:", country)
print("Year recorded:", year)
```

#6.Which country has the lowest life expectancy, and in which year was it recorded?

```python
min_life_expectancy = df.loc[df['Life expectancy'].idxmin()]
country = min_life_expectancy['Country']
year = min_life_expectancy['Year']
print("Country with the lowest life expectancy:", country)
print("Year recorded:", year)
```

#7.What is the correlation between life expectancy and GDP?

```python
correlation = df['Life expectancy'].corr(df['GDP'])
print("Correlation between life expectancy and GDP:", correlation)
```

#8.How does the average life expectancy differ between developed and developing countries?

```python
avg_life_expectancy_by_status = df.groupby('Status')['Life expectancy'].mean()
print("Average life expectancy by country status:")
print(avg_life_expectancy_by_status)
```

#9.What is the distribution of alcohol consumption across the countries?

```python
alcohol_distribution = df.groupby('Country')['Alcohol'].mean()
print("Distribution of alcohol consumption:")
print(alcohol_distribution)
```

#10.How does schooling affect life expectancy?

```python
correlation = df['Schooling'].corr(df['Life expectancy'])
print("Correlation between schooling and life expectancy:", correlation)
```

#11.How many babies deaths were recorded for each country?

```python
infant_deaths_per_country = df.groupby('Country')['infant deaths'].sum()
print("Total infant deaths per country:")
print(infant_deaths_per_country)
```

#12.What is the average percentage expenditure on healthcare across all countries and years?

```python
average_healthcare_expenditure = df['percentage expenditure'].mean()
print("Average percentage expenditure on healthcare:", average_healthcare_expenditure)
```

#13.Relationship between measles incidence and hepatitis B coverage?

```python
correlation = df['Measles '].corr(df['Hepatitis B'])
print("Correlation between measles incidence and hepatitis B coverage:", correlation)
```

#14.What is the distribution of BMI (Body Mass Index) values across the countries?

```python
bmi_distribution = df.groupby('Country')[' BMI '].mean()
print("Distribution of BMI values:")
print(bmi_distribution)
```

#15.How many countries have a life expectancy greater than 75?

```python
countries_gt_75 = df[df['Life expectancy'] > 75]['Country'].nunique()
print("Number of countries with life expectancy greater than 75:", countries_gt_75)
```

#16.What is the maximum alcohol consumption recorded among all countries?

```python
max_alcohol_consumption = df['Alcohol'].max()
print("Maximum alcohol consumption:", max_alcohol_consumption)
```

#17.What is the median GDP per capita?

```python
median_gdp = df['GDP'].median()
print("Median GDP per capita:", median_gdp)
```

#18.What is the average schooling level across all countries?

```python
average_schooling = df['Schooling'].mean()
print("Average schooling level:", average_schooling)
```

#19.What is the average life expectancy for developed and developing countries?

```python
average_life_expectancy_by_status = df.groupby('Status')['Life expectancy'].mean()
print("Average life expectancy by country status:")
print(average_life_expectancy_by_status)
```

#20.Which country has the highest adult mortality rate?

```python
country_highest_adult_mortality = df['Country'].loc[df['Adult Mortality'].idxmax()]
print("Country with the highest adult mortality rate:", country_highest_adult_mortality)
```
